```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

# Set parameters
fs = 30  # Sampling frequency in Hz (you can modify this to observe changes)
duration = 1  # Duration of the signal in seconds
f_signal = 5  # Frequency of the sine wave signal in Hz

# Create the time vector
t = np.linspace(0, duration, fs, endpoint=False)

# Generate a noisy sine wave signal
signal_data = np.sin(2 * np.pi * f_signal * t) + 0.5 * np.random.randn(len(t))

# Compute the periodogram of the signal
frequencies, psd = signal.periodogram(signal_data, fs)

# Plot the signal and its periodogram
plt.figure(figsize=(12, 6))

# Subplot for the signal
plt.subplot(1, 2, 1)
plt.plot(t, signal_data)
plt.title("Signal (Noisy Sine Wave)")
plt.xlabel("Time [s]")
plt.ylabel("Amplitude")

# Subplot for the periodogram
plt.subplot(1, 2, 2)
plt.semilogy(frequencies, psd)  # Use semilogy for better visualization
plt.title("Periodogram (Power Spectral Density)")
plt.xlabel("Frequency [Hz]")
plt.ylabel("Power [dB]")
plt.xlim(0, fs/2)  # Limit the x-axis to the Nyquist frequency
plt.grid(True)

# Display the plot
plt.tight_layout()
plt.show()
```
